```python
movie_count = pd.DataFrame(df.release_year.value_counts(ascending=True))
movie_count = movie_count.reset_index()
```

```python
movie_count.columns = ['year','count']
movie_count.head()
```

```python
movie_count = movie_count.sort_values(by=['year'])
```

```python
movie_count.head()
```

```python
plt.plot(movie_count['year'],movie_count['count'],c="red",marker='|');
plt.xlabel("Years")
plt.ylabel("Movie counts")
plt.title("Released movies annually")
plt.xlim(1960,2015)
plt.xticks(range(1960,2016,5))
plt.legend(['movies'],loc='best')
plt.show()
```

```python
plt.bar(np.arange(5),np.arange(5))
```

```python
top = df.groupby('original_title').sum().sort_values(by="revenue",ascending=False)
top
```

```python
movies = top.head().index
```

```python
movies
```

```python
revenue = top.head().revenue.values
revenue
```

```python
plt.figure(figsize=(10,5))
plt.bar(movies,revenue,color=['blue','g','0.2','red','0.5'])
plt.xlabel("Movie",fontsize=20)
plt.ylabel("Rvenue",fontsize=20)
plt.grid()
plt.title("Best movie's revenue")
```
