```python
ml = pd.read_csv('insurance.csv')
ml.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

#ml = ml.drop('children',axis=1)
#ml = ml.drop('sex',axis=1)
#ml = ml.drop('age',axis=1)
ml = ml.drop('region',axis=1)

ml['smoker'] = le.fit_transform(ml['smoker'])
ml.head(3)
```

```python
sns.heatmap(ml.corr(),annot=True);
```

```python
x = ml.drop('charges',axis=1)
y = ml['charges']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.15,random_state=23)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
print(lm.intercept_)
coeff = pd.DataFrame(lm.coef_,x.columns,columns=['Coeff'])
print(coeff)
p = lm.predict(x_test)
```

```python
plt.figure(figsize=[3,3]);
sns.displot((y_test-p),bins=20)
```

```python
rv = np.array(y_test)
plt.scatter(rv,p)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         'r--', lw=2)
```

```python
from sklearn import metrics as mt
print(
    mt.mean_absolute_error(rv,p),
    mt.mean_squared_error(rv,p),
    mt.root_mean_squared_error(rv,p),
    mt.r2_score(rv,p)
)
```
